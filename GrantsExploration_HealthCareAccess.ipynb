Clean it up and get it into a dataframe

```python
# Reading the entire file and parsing it as JSON
with open('HealthCareAccess.txt', 'r') as file:
    health_care_access_data_content = file.read()

# Parsing the JSON content
health_care_access_data_json = json.loads(health_care_access_data_content)

# Extracting the health care access grant details from the "rows" key
health_care_access_grants_data = health_care_access_data_json["data"]["rows"]

# Converting to a pandas DataFrame
health_care_access_df = pd.DataFrame(health_care_access_grants_data)

# Displaying the top 20 rows of the DataFrame
health_care_access_df.head(20)
```

# Chart the Top 10 Funders by Grant Ammount

```python
# Converting the amount_usd column to float for aggregation
health_care_access_df['amount_usd'] = health_care_access_df['amount_usd'].astype(float)

# Identifying the top 10 funders by total grant amount for health care access grants
health_care_access_top_funders = health_care_access_df.groupby('funder_name')['amount_usd'].sum().nlargest(10)

# Plotting the top 10 funders
health_care_access_top_funders_plot = health_care_access_top_funders.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Funders for Health Care Access Grants')
plt.xlabel('Total Amount (USD)')
plt.ylabel('Funder Name')
plt.gca().invert_yaxis()  # Inverting the y-axis to display the largest value at the top
plt.show()
```

# Lets look at the grant subjects

```python
# Analyzing the top 10 grant subjects by count for health care access grants
health_care_access_top_subjects = health_care_access_df['grant_subject_tran'].value_counts().nlargest(10)

# Plotting the top 10 grant subjects
health_care_access_top_subjects_plot = health_care_access_top_subjects.plot(kind='barh', figsize=(10, 6), color='lightgreen')
plt.title('Top 10 Grant Subjects for Health Care Access Grants')
plt.xlabel('Number of Grants')
plt.ylabel('Subject')
plt.gca().invert_yaxis()  # Inverting the y-axis to display the largest value at the top
plt.show()
```

# We need to break out the individual subjects contained in the rows. They are seperated by semicolons.

```python
# Splitting the 'grant_subject_tran' column by semicolons and expanding into a Series
all_subjects = health_care_access_df['grant_subject_tran'].dropna().str.split(';').explode()

# Counting the occurrences of each unique subject and getting the top 10
top_subjects = all_subjects.value_counts().nlargest(10)

# Plotting the top 10 grant subjects
top_subjects_plot = top_subjects.plot(kind='barh', figsize=(10, 6), color='lightgreen')
plt.title('Top 10 Grant Subjects for Health Care Access Grants')
plt.xlabel('Number of Grants')
plt.ylabel('Subject')
plt.gca().invert_yaxis()  # Inverting the y-axis to display the largest value at the top
plt.show()
```

# Look at the populations most awarded.

```python
# Analyzing the top 10 target populations by count for health care access grants
health_care_access_top_populations = health_care_access_df['grant_population_tran'].value_counts().nlargest(10)

# Plotting the top 10 target populations
health_care_access_top_populations_plot = health_care_access_top_populations.plot(kind='barh', figsize=(10, 6), color='lightcoral')
plt.title('Top 10 Target Populations for Health Care Access Grants')
plt.xlabel('Number of Grants')
plt.ylabel('Target Population')
plt.gca().invert_yaxis()  # Inverting the y-axis to display the largest value at the top
plt.show()
```

Same issue as previously there are multiple populations seperated by semicolons so we will split these targets as well.

```python
# Splitting the 'grant_population_tran' column by semicolons and expanding into a Series
all_populations = health_care_access_df['grant_population_tran'].dropna().str.split(';').explode()

# Counting the occurrences of each unique target population and getting the top 10
top_populations = all_populations.value_counts().nlargest(10)

# Plotting the top 10 target populations
top_populations_plot = top_populations.plot(kind='barh', figsize=(10, 6), color='lightcoral')
plt.title('Top 10 Target Populations for Health Care Access Grants')
plt.xlabel('Number of Grants')
plt.ylabel('Target Population')
plt.gca().invert_yaxis()  # Inverting the y-axis to display the largest value at the top
plt.show()
```

 Now we will look at the distribution by state and see if we spot any trends

```python
# Grouping by state and summing the grant amounts for health care access grants
health_care_access_state_distribution = health_care_access_df.groupby('recip_state')['amount_usd'].sum().nlargest(10)

# Plotting the distribution across states
health_care_access_state_distribution_plot = health_care_access_state_distribution.plot(kind='barh', figsize=(10, 6), color='gold')
plt.title('Top 10 States by Grant Amount for Health Care Access Grants')
plt.xlabel('Total Amount (USD)')
plt.ylabel('State')
plt.gca().invert_yaxis()  # Inverting the y-axis to display the largest value at the top
plt.show()
```

This code will export the HealthcareAccess File to a CSV for excel, we split the grant subjects and populations to generate new rows and give each grant award a grant ID. If you dont want this behavior comment out the lines.

```python
import pandas as pd
import json

# Path to the JSON file
file_path = 'HealthCareAccess.txt'

# Reading the entire file and parsing it as JSON
with open(file_path, 'r') as file:
    data_content = file.read()

# Parsing the JSON content
data_json = json.loads(data_content)

# Extracting the grant details from the "rows" key
grants_data = data_json["data"]["rows"]

# Converting to a pandas DataFrame
df = pd.DataFrame(grants_data)

# Adding a unique grant ID for each original grant record
df['grant_id'] = range(1, len(df) + 1)

# Splitting the 'grant_subject_tran' and 'grant_population_tran' columns by semicolons
df['grant_subject_tran'] = df['grant_subject_tran'].apply(lambda x: str(x).split(';') if x else [])
df['grant_population_tran'] = df['grant_population_tran'].apply(lambda x: str(x).split(';') if x else [])

# Using the 'explode' method to create separate rows for each unique subject and population
df_exploded = df.explode('grant_subject_tran').explode('grant_population_tran')

# Saving to CSV
csv_path = 'HealthCareAccess.csv'
df_exploded.to_csv(csv_path, index=False)
print(f"CSV file saved to {csv_path}")
```

# Lets heatmap the Population and Subject by $$ awarded to see if we can spot anything there.

```python
import plotly.express as px

# Reading the entire file and parsing it as JSON
with open('HealthCareAccess.txt', 'r') as file:
    health_care_access_data_content = file.read()

# Parsing the JSON content
data_json = json.loads(health_care_access_data_content)

# Extracting the grant details from the "rows" key
grants_data = data_json["data"]["rows"]

# Converting to a pandas DataFrame
df = pd.DataFrame(grants_data)

# Splitting the 'grant_subject_tran' and 'grant_population_tran' columns by semicolons
df['grant_subject_tran'] = df['grant_subject_tran'].apply(lambda x: str(x).split(';') if x else [])
df['grant_population_tran'] = df['grant_population_tran'].apply(lambda x: str(x).split(';') if x else [])

# Using the 'explode' method to create separate rows for each unique subject and population
df_exploded_subjects_populations = df.explode('grant_subject_tran').explode('grant_population_tran')

# Creating a pivot table for heatmap
pivot_table_subjects_populations = pd.pivot_table(df_exploded_subjects_populations, values='amount_usd', index='grant_subject_tran', columns='grant_population_tran', aggfunc='sum')

# Creating an interactive heatmap using Plotly Express
fig_heatmap = px.imshow(pivot_table_subjects_populations, title='Health Care Access Grants by Subject and Target Population')
fig_heatmap.show()


```
